#### Get BTC's Daily Simple Return

```python
# Import the libraries
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

%matplotlib inline
```

```python
# Use YFinance to download the data
df = yf.download('BTC-USD', start='2018-01-02')
#df = df.reset_index()
df.head()
```

```python
# Visually show and plot the close price

plt.figure(figsize=(14, 8))
plt.title('Close Price', fontsize=18)
plt.plot(df['Close'])
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price', fontsize=18)
plt.show()
```

```python
# Calculate and show the Daily Simple Return (DSR)
DSR = df['Close'].pct_change()
# Show the daily simple return
DSR
```

```python
# Get some statistics on the Daily Simple Return
DSR.describe()
```

```python
# Avg daily return in %
DSR.mean()*100
```

```python
# Visually show and plot the Daily Simple Return
plt.figure(figsize=(14, 8))
plt.title('Daily Simple Returns', fontsize=18)
plt.plot(DSR.index, DSR, label=DSR, lw=2, alpha=0.7)
plt.xlabel('Date', fontsize=18)
plt.ylabel('Percentage', fontsize=18)
plt.xticks(rotation=45)
plt.show()
```

```python
# Overlay the two charts - was the largest negative return the best time to buy BTC?
# First plot the close price
top = plt.subplot2grid((4,4), (0,0), rowspan=3, colspan=4)
top.plot(df.index, df['Close'], label='Close')
plt.title('Close Price')
plt.legend(loc='upper left')
# Plot the Daily Simple Return
bottom = plt.subplot2grid((4,4), (3,0), rowspan=1, colspan=4)
plt.title('DSR')
bottom.plot(DSR.index, DSR)
plt.subplots_adjust(hspace=0.75)
plt.gcf().set_size_inches(15, 8)
plt.show()
```

### Seems like buying at the biggest day loss would have resulted in a 6X gain in bitcoin by the next year.
