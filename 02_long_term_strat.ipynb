```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Create lists
Year = []
Annual_Income = []
Annual_Expenses = []
Annual_Returns = []
Annual_Investments = []
```

```python
# Yearly income - remains constant
income = 35000
# income = 75000
# Yearly expense - remains constant
expense = income / 2
# Average interest of 8% per year
interest_rate = 0.08
# You invest half your income
investment = income / 2
# Calculating annual return
annual_return = investment * interest_rate
# Getting the current year
year = 2021
```

```python
# Append the first values to the list
Year.append(year)
Annual_Income.append(income)
Annual_Expenses.append(expense)
Annual_Investments.append(investment)
Annual_Returns.append(annual_return)
```

```python
# Loop for n years
invested_years = 20
for i in range(0, invested_years-1):
    # Update the investment to be the previous investment plus the previous annual return plus half of your incomee
    # annual_return = investment * interest_rate
    investment = investment + annual_return + income / 2
    # Update the annual return to be the current investment times the interest rate
    annual_return = investment * interest_rate
    # Append the new data to the list
    Year.append(year+i+1)
    Annual_Income.append(income)
    Annual_Expenses.append(expense)
    Annual_Investments.append(investment)
    Annual_Returns.append(annual_return)
```

```python
# Create a dataframe
df = pd.DataFrame()
df['Year'] = Year
df['Annual_Income'] = Annual_Income
df['Annual_Expenses'] = Annual_Expenses
df['Annual_Investments'] = Annual_Investments
df['Annual_Returns'] = Annual_Returns
```

```python
# Show the data
df
```

```python
# Visually show the data
plt.figure(figsize=(12, 8))
plt.plot(df['Year'],df['Annual_Expenses'], label='Annual Expenses')
plt.plot(df['Year'],df['Annual_Returns'], label='Annual Returns')
plt.xticks(rotation=45)
plt.xticks(df['Year'])
plt.xlabel('Years')
plt.ylabel('$')
plt.title('Retire in 10 Years')
plt.legend()
plt.show()
```
